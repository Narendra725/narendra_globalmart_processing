```python
import json 
with open(r"22d13816-59fd-44d1-97dc-c98af92c480c_83d04ac6-cb74-4a96-a06a-e0d5442aa126_globalmart_data_from_0th_offset.json","r") as file:
    data=json.load(file)
```

```python
import pprint
pprint.pprint(data)
```

```python
mydict={}
for k in data:
    if k["product"]["product_name"] in mydict.keys():
        if type(k["product"]["sizes"]) != "null" :
            mydict[k["product"]["product_name"]]["sizes"]=list(set(mydict[k["product"]["product_name"]]["sizes"]+list(k["product"]["sizes"].split(","))))
            mydict[k["product"]["product_name"]]["counts"]=len(mydict[k["product"]["product_name"]]["sizes"])
    else:
        if type(k["product"]["sizes"]) != "null" :
            mydict[k["product"]["product_name"]]={"sizes":list(set(k["product"]["sizes"].split(",")))}
            mydict[k["product"]["product_name"]]["counts"]=len(mydict[k["product"]["product_name"]]["sizes"])
        
```

```python
{k: v for k, v in sorted(mydict.items(), key=lambda item: item[1]["counts"],reverse =True)}
```
